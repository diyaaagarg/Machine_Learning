```python
import seaborn as sns
```

```python
import pandas as pd
tips=pd.read_csv('tips.csv')
```

```python
##basic plotting with seaborn
sns.load_dataset('tips')
tips
```

```python
##create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips,marker='*')
plt.title("Scatter plot of total bill verses tip")
plt.show()
```

```python
#LINE PLOT
sns.lineplot(x="size",y='total_bill',data=tips)
plt.title("line plot of size verses total bill")
plt.show()
```

```python
#categorical plots
#bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("bar plot of day verses total bill")
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("box plot of day verses total bill")
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("violin plot of day verses total bill")
```

```python
##histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#in case kde=false,curve will not show
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

```python
#KDE PLOT
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
#in case of shade=false,shading part will remove
sns.kdeplot(tips['total_bill'],shade=False)
```

```python
#pairplot
#If I want to see the relationship between all the variable to different different variables
sns.pairplot(tips)
```

```python
#HEATMAP
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
