```python
import numpy as np
import pandas as pd
```

```python
#set random seeds for reproducibility
np.random.seed(123)
#create a dataframe with two classes
n_samples=1000
class_0_ratio=0.9
n_class_0=int (n_samples*class_0_ratio)
n_class_1=int (n_samples-n_class_0)
```

```python
n_class_0,n_class_1
```

```python
#CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
#UPSAMPLING  #increase minority wala
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,  #sample with replacement
         n_samples=len(df_majority),
         random_state=42
         )
df_minority_upsampled.shape  #increased minority counts
```

```python
df_upsampled=pd.concat([df_minority_upsampled,df_majority])
```

```python
df_upsampled['target'].value_counts()
```

```python
#DOWNSAMPLEING
#CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
df=pd.concat([class_0,class_1]).reset_index(drop=True)
df['target'].value_counts()
```

```python
#DOWNSAMPLING  #decrease majority wala
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_majority_downsampled=resample(df_majority,replace=False,  #sample with replacement
         n_samples=len(df_minority),
         random_state=42
         )
df_majority_downsampled.shape  #increased minority counts
```

```python
df_downsampled=pd.concat([df_minority,df_majority_downsampled])
```

```python
df_downsampled['target'].value_counts()
```
