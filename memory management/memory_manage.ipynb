#Reference Counting
# it is a primary method Python uses to manage memory. Each object in Python maintains a count of references pointing to it.
# When the references count drops to zero, the memory occupied by the object is deallocated.

```python
import sys
a=[]
#o/p=2 beacause one reference coming from 'a' and one from getrefcount function
print(sys.getrefcount(a))
```

```python
b=a 
#o/p=3 because one from 'a' ,one from 'b' and one from getrefcount
print(sys.getrefcount(b))
```

```python
del b
print(sys.getrefcount(a))
```

#GARBAGE COLLECTION
Python includes a cyclic garbage collector to handle reference cycles. Reference cycles occurs when objects reference each other, preventing the reference counts from reaching zero

```python
import gc
#enable garbage collection
gc.enable()
```

```python
gc.disable()
```

```python
gc.collect() 
# O/P-> Number of unreachable objects 
```

```python
#get garbage collection stats
print(gc.get_stats())
```

```python
#get unreachable objects
print(gc.garbage)
```

MEMORY MANAGEMENT BEST PRACTICES
1)USE LOCAL VARIABLES:Local variable have a shorter lifespan and freed sooner then global variables.
2)AVOID CIRCULAR REFERENCES:circular reference can lead to memory leaks IF NOT PROPERLY MANAGED
3)USE GENERATORS:generator produce items one at a time and only keep one item in memory at a time,making them memory efficient
4)EXPLICITY DELETE OBJECTS:Use the delete statement to delete variable and objects explicitly
5)PROFILE MEMORY USAGE:using the memory profiling tools like trace, malloc and Memory Profiler to identify memory leaks and optimize memory usage.

```python
#HANDLED CIRCULAR REFERENCE
import gc
class MyObject:
    def __init__(self,name):
        self.name=name
        print(f"Object {self.name}created")

    def __del__(self):
        print(f"object{self.name}deleted")

#create circular reference
obj1=MyObject("obj1")
obj2=MyObject("obj2")
obj1.ref=obj2
obj2.ref=obj1

del obj1
del obj2

#after using del, both the objects will be unreachable.But this will not be garbage collected because due to circular reference we have the 
# circular reference of object one with object two, object two with object one.

# So for this particular case,we need to manually trigger the garbage collection
gc.collect()
```

```python
#GENERATORS FOR MEMORY EFFICIENCY
#generator allows you to produce one item at a time using memory efficiently by only keeping one item in the memory and removing that particular
#thing.
def generate_numbers(n):
    for i in range (n):
        yield i
#using the generator
for num in generate_numbers(100000):
    print(num)
    if num>10:
        break
```

```python
#profiling memory usage with tracemalloc
import tracemalloc

def create_list():
    return [i for i in range(10000)]
def main():
    tracemalloc.start()
    create_list()
    snapshot=tracemalloc.take_snapshot()
    top_stats=snapshot.statistics('lineno')
    print("[Top 10]")
    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
