PYTHON LOGGING
logging is crucial aspects of any application providing ways to track event errors and operational information.
Python built in logging module offers a flexible framework for emitting log messages for Python program.
In this lesson, we'll cover the basics of logging, including how to configure logging, logging levels,and the best practices for using logging in Python application.

```python
import logging

#configure basic logging settings
logging.basicConfig(level=logging.DEBUG)

#log messeges with different security levels
logging.debug("this is a debug messege")
logging.info("this is an info messge")
logging.warning("this is a warning messge")
logging.error("this is a error messge")
logging.critical("this is a critical messege")
```

LOG LEVELS:
Python logging modules has several log levels indicating the severity of events.The default levels are:
->debug: detailed information, typically of interest only when diagnosis problem.
->info: confirmation that things are working as expected.
->warning: an indication that something unexpected happened or indicative of some problem in the near future.(eg->disc space low).the sopftware is still working as expected
->error: due to your most serious problem, the software has not been able to perform some function
->critical:A very serious error, indicating the program itself may be unable to continue running.

```python
#configuring logging
import logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d%H:%M:%S'
)
#log messeges with different security levels
logging.debug("this is a debug messege")
logging.info("this is an info messge")
logging.warning("this is a warning messge")
logging.error("this is a error messge")
logging.critical("this is a critical messege")
```
