LOGGING WITH MULTIPLE LOGGERS
->you can create multiple loggers for different parts of your application


```python
import logging
#create a logger for module1
logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

#create a logger for module2
logger2=logging.getLogger("modeule2")
logger2.setLevel(logging.WARNING)

#configure logging settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d%H:%M:%S'
)

```

```python
#log messeges with different loggers
logger1.debug("this is debug messege for module1")
logger2.warning("this is warning messege for module2")
logger2.error("this is error messege for module2")
```
