```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)#header is used to merge both partition in dataset
```

```python
dataset.info()
```

# DATA CLEANING

```python
#missing values
dataset[dataset.isnull().any(axis=1)]
```

the dataset is converted into two sets based on region from 122th index,we can make a new column based on the region
1:bejaia region dataset
2:Sidi-Bel Abbes Region dataset
add new column with region 

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
#removing the null values
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fix spaces in column name
df.columns=df.columns.str.strip()
df.columns
```

CHANGES THE REQUIRED COLUMNS AS INTEGER DATA TYPE

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
#changing the other columns as float datatype
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

# LET SAVE THE CLEANED CSV FILE

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
## plot density plot for all features
import seaborn as sns
sns.set_theme()
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
#percentage for pie chart
df_copy['Classes'].value_counts(normalize=True)  #normalize=True:convert in percentage
```

```python
#percentage of pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#plotting piechart
classlabels=["Fire","Not fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Distribution of Fire vs Not Fire")
plt.show()
```
