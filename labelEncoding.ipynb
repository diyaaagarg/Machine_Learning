```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'color':['red','blue','green','green','red']
})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

ORDINAL ENCODING 

In this technique, each category is assigned a numerical value based on its position in the order.
For example, if we have a categorical variable education level with four possible values like high
school, college graduate, and post graduate, we can represent it using ordinal encoding.:
1.HIGH SCOOL:1
2.COLLEGE:2
3.GRADUATE:3
4.POST-GRADUATE:4

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    'size':['small','large','medium','small','large']
})
df
```

```python
#create an instance of ordinalEncoder and then fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```
